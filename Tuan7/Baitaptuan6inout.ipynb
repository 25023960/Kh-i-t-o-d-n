## W6A9

```python
def ipd(a):
    c={}
    for i in a:
        key,value=i.split(":")
        c[key]=c.get(key,0)+int(value)
    return c
a=input().split()
dict1=ipd(a)
b=input().split()
dict2=ipd(b)
for i in dict2:
    dict1[i]=dict1.get(i,0)+dict2[i]
k=dict1.keys()
k=sorted(k)
print({i:dict1[i] for i in k})
```

## W6A10

```python
a=list(map(int,input().split()))
k=int(input())
b=set()
c=set()
for i in a:
    target=k-i
    if target in b:
        c.add((min(i,target),max(i,target)))
    b.add(i)
print(sorted(list(c)))
```

## W6A15

```python
def ipd(a):
    c={}
    for i in a:
        key,value=i.split(":")
        c[key]=c.get(key,0)+int(value)
    return c
a=input().split()
dict1=ipd(a)
k=int(input())
c={}
for i,v in dict1.items():
    if int(v)>k:
        c[i]=v
print(c)
```
