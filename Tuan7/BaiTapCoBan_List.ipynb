### Bài 2: Tìm số nhỏ nhất của một danh sách các số nguyên

```python
# lst = list(map(int, input().split()))

def min_of_list (lst):
    if lst:
        M = lst[0]
        for x in lst:
            if x < M:
                M = x
        return M
    return None

assert min_of_list([]) == None
assert min_of_list([23, 67, 900, -10]) == -10
assert min_of_list([575, 939, 697, 138, 183, 386, 147, 695, 622, 289]) == 138

print("Pass all Tests!")
```

### Bài 2b: Tìm số nhỏ nhất của 1 list và vị trí của nó, nếu có nhiều số nhỏ nhất bằng nhau thì lấy chỉ số đầu tiên (nhỏ nhất)
10 0 0 12 34 5 6 7  8 89 0 1
0 1

```python
#Khoi

def min_and_pos(lst):
    if lst:
        min_value = lst[0]
        min_pos = 1
        for i, x in enumerate(lst, 1):
            if x < min_value:
                min_value = x
                min_pos = i
        return min_value, min_pos
    return None

assert min_and_pos([10, 0, 0, 12, 34, 5, 6, 7, 8, 89, 0, 1]) == (0, 2)
assert min_and_pos([]) == None

print("Pass all Tests!!!")
```

### Bài 3: 
Tìm giá trị trung bình của các số nguyên dương trong một list.
Nếu không có số dương nào thì trả về "invalid"

```python
#Tuan
def Average_lst(lst):
    if lst:
        t = 0
        d = 0
        for x in lst:
            if x > 0:
                t += x
                d += 1
        if not d:
            return "invalid"
        return round(t/d, 2)
    
    return "invalid"

assert Average_lst([1, 2, 6, 1, -10, -10]) == 2.50
assert Average_lst([-2, -1, -20, -343]) == "invalid"
assert Average_lst([]) == "invalid"

print("Accepted!")
```

### Bài 5: Đưa tất cả các số bằng 0 trong list về đầu của list. Các số khác không thay đổi thứ tự trước sau
3 4 6 7 1 9 0 0 23 0 0 34 67 89
0 0 0 0 3 4 6 7 1 9 23 34 67 89

```python
def move_zero_to_first(lst):
    if lst:
        lst_zero = []
        lst_non_zero = []
        for x in lst:
            if x == 0:
                lst_zero.append(x)
            else:
                lst_non_zero.append(x)
        return lst_zero + lst_non_zero
    return None

assert move_zero_to_first([3, 4, 0, 0, 23, 0, 0, 34, 67, 89]) == [0, 0, 0, 0, 3, 4, 23, 34, 67, 89]
assert move_zero_to_first([0, 0, 0, 0]) == [0, 0, 0, 0]

print("accepted!!!")
```

### Bài 6: Xóa tất cả các số giống nhau trong list, Các số khác giữ nguyên vị trí trước 
[1, 2, 3, 4, 1, 2, 3, 4, 1, 1, 4, 3] 


```python
def unique_array(lst):
    if lst:
        ans = []
        for x in lst:
            if x not in ans:
                ans.append(x)
        return ans
    return None


assert unique_array([1, 2, 3, 4, 1, 2, 3, 4, 1, 1, 4, 3]) == [1,2,3,4]

print("Accepted!!!")
```

### Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang

9 8 7 6 5 4 0: Giam

1 2 3 7 1 0 2: Vo danh

1 2 2 4 5 7: Vo danh

```python
a=list(map(int,input().split()))
if a[1]>a[0]:
    k="Tang"
elif a[1]<a[0]:
    k="Giam"
else:
    k="Vo danh"
for i in range(len(a)-1):
    if k=="Tang" and a[i]>=a[i+1]:
        k="Vo danh"
        break
    elif k=="Giam" and a[i]>=a[i+1]:
        k="Vo danh"
        break
print(k)
```

### Bài 9: Tìm và in ra tất cả các vị trí của số nguyên k trong một danh sách số nguyên cho trước. Nếu không có k trong ds thì in ra "-1".
VD: k = 3, [1,3,4,4,3,3,4,3,6,1] => k xuất hiện tại các vị trí (trả về 1 list) [1, 4, 5, 7]
k = 10, [1,3,4,4,3,3,4,3,6,1] => k không xuất hiện trong ds => trả về "-1"

```python
def i(a,k):
    b=[]
    for i,char in enumerate(a):
        if char==k:
            b.append(i)
    if b==[]:
        return -1
    return b
a=list(map(int,input().split()))
k=int(input())
print(i(a,k))
```

### Bài 11: 
Cho một danh sách các số nguyên có giá trị trong đoạn [0,9]. Liệt kê số lần xuất hiện của các số có mặt trong danh sách. Số nào nhỏ thì liệt kê trước
VD 1 2 1 2 1 1 1 1

output:

```python
a=list(map(int,input().split()))
c={}
a.sort()
for i in a:
    c[i]=c.get(i,0)+1
print(c)
```

### Bài 12: 
Cho một chuỗi kí tự chỉ gồm các kí tự chữ cái thường 'a'....'z'. Hãy liệt kê số lần xuất hiện của các kí tự có mặt trong chuỗi theo thứ tự tăng dần.
VD: "abcabcccc"
output:
a : 2
b : 2
c : 5

```python
a=str(input())
c={}
for i in a:
    c[i]=c.get(i,0)+1
for i in sorted(c.keys()):
    print(f"{i}:{c[i]}",end=" ")
```
