## W6A9

```python
def ipd(a):
    c={}
    for i in a:
        key,value=i.split(":")
        c[key]=c.get(key,0)+int(value)
    return c
a=input().split()
dict1=ipd(a)
b=input().split()
dict2=ipd(b)
for i in dict2:
    dict1[i]=dict1.get(i,0)+dict2[i]
k=dict1.keys()
k=sorted(k)
print({i:dict1[i] for i in k})
```

## W6A10

### Two pointers

```python
a=list(map(int,input().split()))
k=int(input())
b=sorted(set(a))
l=0
r=len(b)-1
li=[]
while l<r:
    if b[l]+b[r]==k:
        li.append((b[l],b[r]))
        l+=1
        r-=1
    elif b[l]+b[r]<k:
        l+=1
    else:
        r-=1
if k%2==0 and k//2 in b and a.count(k//2)>1:
    li.append((k//2,k//2))
print(li)
```

### Hash map

```python
a=list(map(int,input().split()))
k=int(input())
b=set()
c=set()
for i in a:
    target=k-i
    if target in b:
        c.add((min(i,target),max(i,target)))
    b.add(i)
print(sorted(list(c)))
```
