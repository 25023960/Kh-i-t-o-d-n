## W6A16

```python
n,m=map(int,input().split())
k=[]
for i in range(n):
    row=list(map(int,input().split()))
    if len(row)!=m:
        print("Loi")
        break
    k.append(row)
for row in k:
    for num in row:
        print(f"{num:>4}",end="")
    print()
```

## W6A17

```python
n=int(input())
c=0
d=n-1
k=[]
l1=[]
l2=[]
for i in range(n):
    row=list(map(int,input().split()))
    k.append(row)
for row in k:
    l1.append(row[c])
    c+=1
for row in k:
    l2.append(row[d])
    d-=1
print(l1)
print(l2)
```

## W6A18

```python
n,m,k=map(int,input().split())
l=[]
for i in range(n):
    row=list(map(int,input().split()))
    if len(row)!=m:
        print("LOI")
        break
    l.append(row)
cot=0
for row in l:
    cot+=row[k-1]
hang=0
for i in l[k-1]:
    hang+=i
print(cot)
print(hang)
```

## W6A26

```python
a=list(map(int,input().split()))
tong=0
if a[0]<a[1]:
    k="Tang"
if a[0]>a[1]:
    k="Giam"
for i in range(1,len(a)-1):
    if a[i]>a[i+1] and k=="Tang":
        tong+=a[i]
        k="Giam"
    elif a[i]<a[i+1] and k=="Giam":
        k="Tang"
print(tong)
```

## W6A27

```python
a=list(map(int,input().split()))
x=int(input())
k=[]
for i in range(len(a)):
    if a[i]<x:
        k.append(a[i])
if k==[]:
    print(-1)
else:
    print(*k)
```

```python
a=list(map(int,input().split()))
if a[1]>a[0]:
    k="Tang"
elif a[1]<a[0]:
    k="Giam"
else:
    k="Vo danh"
for i in range(2,len(a)-1):
    if k=="Tang" and a[i]>a[i+1]:
        k="Vo danh"
        break
    elif k=="Giam" and a[i]<a[i+1]:
        k="Vo danh"
        break
if k=="Tang":
    print("Tang thuc su")
elif k=="Giam":
    print("Giam thuc su")
else:
    print('Vo danh')
```
