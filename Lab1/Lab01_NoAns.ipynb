### Problem 4
#### Write a function that returns the count of positive integers k such that 1 <= k < n and k > (n - 1) / 2. (Example: n=9 -> 4).

```python
n=int(input())
def duong(n):
    m=0
    for i in range(2,n+1,2):
        m+=1
    print(m)
duong(n)
```

### Problem 6
#### Write a function that computes the area of a triangle with side lengths a, b, c (use Heron's formula). Raise ValueError if the sides cannot form a triangle.

```python
a,b,c=map(int,input().split())
def dientich(a,b,c):
    if a+b>c and a+c>b and b+c>a:
        p=(a+b+c)/2
        print((p*(p-a)*(p-b)*(p-c))**0.5)
    else:
        print("ValueError")
dientich(a,b,c)
```

### Problem 7

### Problem 8
#### Write a function that returns the character that is n positions after character x in the alphabet.
#### Preserve case and wrap around within A..Z or a..z.

```python
def sau(x,n):
    if "a"<=x<="z":
        k=ord("a")
    elif "A"<= x<="Z":
        k=ord("A")
    else:
        return False
    return chr((ord(x)-k+n)%26+k)
x=input()
n=int(input())
sau(x,n)
```

### Problem 9
#### Write a function that converts an uppercase letter to lowercase. If the input is not a single character, raise ValueError.

```python
def bien(a):
    if len(a)>1:
        return ValueError
    else:
        return a.lower()
a=input()
bien(a)
```

### Problem 10
#### Write a function that, given a lowercase letter, returns a string containing that letter repeated 10 times.

```python
def lap(a):
    if a.islower():
        return a*10
    else:
        return False
a=input()
lap(a)
```

### Problem 11
#### Write a function that returns a right isosceles triangle of height n using '*'.
#### Line i (1-indexed) has i '*', joined with newlines.
#### Example with n = 4
#### *
#### **
#### ***
#### ****

```python
n=int(input())
for i in range(1,n+1):
    print("*"*i)
```
