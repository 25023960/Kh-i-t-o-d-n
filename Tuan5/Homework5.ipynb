### Toi uu

```python
a,b=list(map(int,input().split()))
while b!=0:
    a,b=b,a%b
k=[]
for i in range(1,int(a**0.5)+1):
    if a%i==0:
        k.append(i)
        if i!=a//i:
            k.append(a//i)
k.sort()
print(*k)
            
```
