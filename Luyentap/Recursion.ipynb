## B1

```python
def gt(a):
    if a==0:
        return 1
    else:
        c=gt(a-1)
    return a*c
a=int(input())
print(gt(a))
```

## B2

```python
def bum(a):
    if a==0:
        return "Bum"
    else:
        print(a)
        return bum(a-1)
a=int(input())
print(bum(a))
```

## B3

```python

def tong(a):
    if a==[]:
        return 0
    else:
        return a[0]+tong(a[1:])
a=list(map(int,input().split()))
print(tong(a))
```

## B4

```python
def dx(a):
    if len(a)==1 or a==[]:
        return True
    else:
        if a[0]==a[-1]:
            return dx(a[1:-1])
        return False
a=list(map(int,input().split()))
print(dx(a))
```

## B5


```python
def bi(a,target,l,r):
    if l>r:
        return -1
    mid=(l+r)//2
    if a[mid]==target:
        return mid
    elif a[mid] < target:
        return bi(a,target,mid+1,r)
    else:
        return bi(a,target,l,mid-1)
    
a=list(map(int,input().split()))
target=int(input())
l,r=0,len(a)-1
print(bi(a,target,l,r))
```
