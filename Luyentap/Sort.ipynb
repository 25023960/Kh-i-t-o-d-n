## Binary search

```python
a=list(map(int,input().split()))
k=int(input())
l=0
r=len(a)-1
while l<=r:
    mid=(l+r)//2
    if a[mid]==k:
        print(mid)
        break
    elif a[mid]<k:
        l=mid+1
    else:
        r=mid-1
```

## Bubble sort


```python
a=list(map(int,input().split()))
def bb(a):
    l=len(a)
    for i in range(l):
        s=False
        for j in range(l-i-1):
            if a[j]>a[j+1]:
                a[j],a[j+1]=a[j+1],a[j]
                s=True
        if not s:
            return a
print(bb(a))
```

## Selection sort

```python
a=list(map(int,input().split()))
def sl(a):
    l=len(a)
    for i in range(l):
        mi=i
        for j in range(i+1,l):
            if a[j]<a[mi]:
                mi=j
        a[i],a[mi]=a[mi],a[i]
    return a
print(sl(a))
```

## Insertion sort

```python
a=list(map(int,input().split()))
def ins(a):
    l=len(a)
    for i in range(1,l):
        r=a[i]
        j=i-1
        while j>=0 and r<a[j]:
            a[j+1]=a[j]
            j-=1
        a[j+1]=r
    return a
print(ins(a))
```

## Merge sort


```python
def tach(a):
    if len(a)<=1:
        return a
    mid=len(a)//2
    l=a[:mid]
    r=a[mid:]
    p1=tach(l)
    p2=tach(r)
    return sx(p1,p2)
def sx(p1,p2):
    i=0
    j=0
    k=[]
    while i<len(p1) and j<len(p2):
        if p1[i]<p2[j]:
            k.append(p1[i])
            i+=1
        else:
            k.append(p2[j])
            j+=1
    k.extend(p1[i:])
    k.extend(p2[j:])
    return k
a=list(map(int,input().split()))
print(tach(a))
```
