```python
def tien(car,start,km,wait):
    k=0
    if wait>5:
        k+=(wait//5)*2000
    if 0<km<0.8:
        if car==4:
            k+=11000
        elif car==7:
            k+=12000
    elif 0.8<=km<=20:
        if car==4:
            k+=11000+13000*(km-0.8)
        elif car==7:
            k+=12000+14000*(km-0.8)
    else:
        if car==4:
            k+=11000+13000*(19.2)+10500*(km-20)
        elif car==7:
            k+=12000+14000*19.2+12000*(km-20)
    if start in [22,23,0,1,2,3,4,5]:
        k=k*1.2
    else:
        if km>30:
            k*=0.9
    if k<50 and km>5:
        k=50000
    return int(k)
car,start,km,wait=map(float,input().split())
print(tien(car,start,km,wait))


```

## W25

```python
def point_in_quadrilateral(quad, p):
    n = len(quad)
    positive = False
    negative = False
    for i in range(n):
        x1, y1 = quad[i]
        x2, y2 = quad[(i + 1) % n]
        xp, yp = p
        cross_product = (x2 - x1) * (yp - y1) - (y2 - y1) * (xp - x1)
        if cross_product > 0:
            positive = True
        elif cross_product < 0:
            negative = True
        if positive and negative:
            return False
    return True
```
