## W5A3

```python
def nt(n):
    if n==0 or n==1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
n=int(input())
if nt(n):
    print("YES")
else:
    print("NO")
```

## W5A4

```python
def hh(n):
    k=0
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            k+=i
            if i!=n//i:
                k+=n//i
    if n==k+1:
        return True
    return False
n=int(input())
if hh(n):
    print("YES")
else:
    print("No")
```

## W5A5

```python
def tim(n,k):
    for i in range(len(n)):
        if n[i]==k:
            return i+1
    return -1
n=list(map(int,input().split()))
k=int(input())
print(tim(n,k))
```

## W5A6

```python
def gt(n):
    k=1
    if n==0 or n==1:
        return 1
    for i in range(2,n+1):
        k*=i
    return k
n=int(input())
print(gt(n))
```

## W5A7

```python
def mt(a,x,b):
    if x=="+":
        return a+b
    elif x=="-":
        return a-b
    elif x=="*":
        return a*b
    else:
        return a/b
k=input().split()
a=float(k[0])
x=k[1]
b=float(k[2])
print("{:.2f}".format(mt(a,x,b)))
```

```python
def tong(x):
    k=0
    while x!=0:
        k+=x%10
        x//=10
    return k
x=int(input())
print(tong(x))
```

## W5A10

```python
def dc(a,b):
    x={}
    y={}
    for i,j in zip(a,b):
        if i in x and x[i]!=j:
            return False
        if j in y and y[j]!=i:
            return False
        x[i]=j
        y[j]=i
    return True
a,b=input().split()
if dc(a,b):
    print("True")
else:
    print("False")
```

## W5A11

```python
def ch(a,b):
    if a%10==0 or b%10==0:
        return True
    if a%2==0 and b%5==0:
        return True
    if a%5==0 and b%2==0:
        return True
    return False
a,b=map(int,input().split())
if ch(a,b):
    print("True")
else:
    print("False")
```
