## W5A28

```python
import math
def goc(x1,y1,x2,y2,x3,y3):
    a=x2-x1
    b=y2-y1
    c=x3-x2
    d=y3-y2
    tch=a*c+b*d
    t=(a**2+b**2)**0.5+(c**2+d**2)**0.5
    cosr=tch/t
    r=math.acos(cosr)
    d=math.degrees(r)
    return d
x1,y1,x2,y2,x3,y3=map(int,input().split())
print(goc(x1,y1,x2,y2,x3,y3))
```

## W5A30

```python
def nt(n,k):
    n1=1
    k1=1
    a=1
    for i in range(1,n+1):
        n1*=i
    for i in range(1,k+1):
        k1*=i
    for i in range(1,n-k+1):
        a*=i
    return (n1/(k1*a))
n,k=map(int,input().split())
print(nt(n,k))
```

```python
def tien(car,start,km,wait):
    k=0
    if wait>5:
        k+=(wait//5)*2000
    if 0<km<0.8:
        if car==4:
            k+=11000
        elif car==7:
            k+=12000
    elif 0.8<=km<=20:
        if car==4:
            k+=11000+13000*(km-0.8)
        elif car==7:
            k+=12000+14000*(km-0.8)
    else:
        if car==4:
            k+=11000+13000*(19.2)+10500*(km-20)
        elif car==7:
            k+=12000+14000*19.2+12000*(km-20)
    if start in [22,23,0,1,2,3,4,5]:
        k=k*1.2
    else:
        if km>30:
            k*=0.9
    if k<50 and km>5:
        k=50000
    return int(k)
car,start,km,wait=map(float,input().split())
print(tien(car,start,km,wait))
```

## W5A34


```python
def kh(n):
    a=list(str(n))
    if len(a)==len(set(a)):
        return True
    return False
k=0
n=int(input())
if n<=0:
    print(0)
else:
    for i in range(1,n+1):
        if kh(i):
            k+=i
    print(k)
```
