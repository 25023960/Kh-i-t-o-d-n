## W9A1

```python
def merge(a,b):
    i,j=0,0
    k=[]
    while i<len(a) and j<len(b):
        if a[i]<b[j]:
            k.append(a[i])
            i+=1
        else:
            k.append(b[j])
            j+=1
    k.extend(a[i:])
    k.extend(b[j:])
    return k
a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(merge(a,b))
```

## W9A2

```python
def dem_so(matrix,target):
    n=len(matrix)
    m=len(matrix[0])
    c=0
    row=n-1
    col=0
    while row>=0 and col<m:
        if matrix[row][col]<=target:
            c+=row+1
            col+=1
        else:
            row-=1
    return c
def so_thu(matrix,k):
    n=len(matrix)
    m=len(matrix[0])
    target=m*n-k+1
    l=matrix[0][0]
    r=matrix[n-1][m-1]
    while l<=r:
        mid=(l+r)//2
        if dem_so(matrix,mid)>=target:
            kq=mid
            r=mid-1
        else:
            l=mid+1
    return kq
a=eval(input())
k=int(input())
print(so_thu(a,k))
```

## W9A3

```python
def hd2(a):
    s=[int(i) for i in str(a)]
    l=len(s)
    if l<=1:
        return a
    for i in range(l):
        mi=i
        for j in range(l-1,i,-1):
            if s[j]<s[mi]:
                mi=j
        if mi!=i:
            s[i],s[mi]=s[mi],s[i]
            return int("".join(map(str,s)))
    s[-1],s[-2]=s[-2],s[-1]
    return int("".join(map(str,s)))
a=int(input())
print(hd2(a))
```

## W9A4

```python
def xau(a,b):
    n=len(a)
    m=len(b)
    dp=[[False]*(m+1) for _ in range(n+1)]
    dp[0][0]=True
    for i in range(n):
        for j in range(m+1):
            if not dp[i][j]:
                continue
            if a[i].islower():
                dp[i+1][j]=True
            if j<m and a[i].upper()==b[j]:
                dp[i+1][j+1]=True
    return dp[n][m]
n=int(input())
for i in range(n):
    a=input()
    b=input()
    print(f"Test thu {i+1}")
    if xau(a,b):
        print("Yes")
    else:
        print("No")
```

```python
def hv(a):
    l=len(a)
    i=l-2
    while i>=0 and a[i]>a[i+1]:
        i-=1
    j=l-1
    if i>=0:
        while a[i]>a[j]:
            j-=1
        a[i],a[j]=a[j],a[i]
    left=i+1
    right=l-1
    while left<right:
        a[left],a[right]=a[right],a[left]
        left+=1
        right-=1
    return a
n=int(input())
a=list(map(int,input().split()))
print(hv(a))
```

## W9A7

```python
def PlaceFlowers(flowerbed,k):
    i=0
    while i<len(flowerbed):
        if flowerbed[i]==1:
            i+=2
        else:
            if i==len(flowerbed)-1 or flowerbed[i+1]==0:
                k-=1
                i+=2
        if k==0:
            return True
    return k==0
flowerbed=list(map(int,input().split()))
k=int(input())
print(PlaceFlowers(flowerbed,k))
```

## W9A8

```python
def is_path_crossing(moves):
    x,y=0,0
    c={(x,y)}
    for i in moves:
        if i=="R":
            x+=1
        elif i=="L":
            x-=1
        elif i=="U":
            y+=1
        elif i=="D":
            y-=1
        if (x,y) in c:
            return True
        c.add((x,y))
    return False
a=input()
print(is_path_crossing(a))
```

```python
def Median(a):
    def sortF(i,j):
        if a[i]>a[j]:
            a[i],a[j]=a[j],a[i]
    sortF(0,1)
    sortF(2,3)
    if a[0]>a[2]:
        a[0],a[2]=a[2],a[0]
        a[1],a[3]=a[3],a[1]
    a[0]=a[4]
    sortF(0,1)
    if a[0]>a[2]:
        return a[0] if a[0]<a[3] else a[3]
    else:
        return a[2] if a[2]<a[1] else a[1]
a=list(map(int,input().split()))
print(Median(a))
```

```python
def power(base,exp):
    if exp==0:
        return 1
    return base*power(base,exp-1)
base,exp=map(int,input().split())
if base==0:
    print(0)
else:
    print(power(base,exp))
```

## W9A14

```python
def tim(a,x):
    k=[]
    c={}
    for i in a:
        c[i]=c.get(i,0)+1
    for id,value in c.items():
        if value>x:
            k.append(id)
    return k
a=list(map(int,input().split()))
x=int(input())
if tim(a,x)==[]:
    print("INVALID")
else:
    print(tim(a,x))
```
