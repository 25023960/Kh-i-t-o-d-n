```python
class Cylinder:
    def __init__(self,r,h):
        self.bk=r
        self.cc=h
    def sbm(self):
        return 2*3.14*(self.bk)**2+2*3.14*(self.bk)*(self.cc)
    def tt(self):
        return (self.bk)**2*3.14*(self.cc)
    def __str__(self):
        return f"{self.bk},{self.cc}"
r,h=map(float,input().split())
cy1=Cylinder(r,h)
print(f"{cy1.sbm():.2f}")
print(f"{cy1.tt():.2f}")
print(cy1.cc)
```

## W8A2

```python
class MyDate:
    def __init__(self,dd,mm,yy):
        self.d=dd
        self.m=mm
        self.y=yy
    def cV(self):
        maxday=-1
        match self.m:
            case 1|3|5|7|8|10|12:
                maxday=31
            case 4|6|8|10|12:
                maxday=30
            case 2:
                if (self.y%4==0 and self.y%100!=0) or (self.y%400==0):
                    maxday=29
                else:
                    maxday=28
        return 1<=self.d<=maxday and self.y>0
    def nextDay(self):
        if (self.cV()):
            day1=MyDate(self.d+1,self.m,self.y)
            if day1.cV():
                return day1
            day1=MyDate(1,self.m+1,self.y)
            if day1.cV():
                return day1
            day1=MyDate(1,1,self.y+1)
            if day1.cV():
                return day1
    def __str__(self):
        return f"{self.d:02d}/{self.m:02d}/{self.y}"
d,m,y=map(int,input().split("/"))
day1=MyDate(d,m,y)
day2=day1.nextDay()
print(day2)
```

```python
class Calculator:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def add(self):
        return self.a+self.b
    def subtract(self):
        return self.a-self.b
    def product(self):
        return self.a*self.b
    def divide(self):
        if self.b==0:
            return "Cannot divide"
        return self.a/self.b
    def power(self):
        return self.a**self.b
    def mod(self):
        return self.a%self.b
    def set_numbers(self,a,b):
        self.a=a
        self.b=b
a,b=map(float,input().split())
kq=Calculator(a,b)
while True:
    print("1. Add | 2. Subtract | 3. Product | 4. Divide | 5. Power | 6. Mod | 7. Set Numbers")
    n=int(input())
    if n==1:
        print(kq.add())
    elif n==2:
        print(kq.subtract())
    elif n==3:
        print(kq.product())
    elif n==4:
        print(kq.divide())
    elif n==5:
        print(kq.power())
    elif n==6:
        print(kq.mod())
    elif n==7:
        a,b=map(float,input().split())
        kq.set_numbers(a,b)
    e=input("Ban co muon thoat ko?(yes/no)").strip().lower()
    if e=="yes":
        print("Bye")
        break
```

```python
class ShoppingCart:
    def __init__(self):
        self.k=[]
    def add(self,name,price):
        self.k.append({"name":name,"price":price})
        print(f"Da them {name}")
    def de(self,name):
        for i in self.k:
            if i["name"]==name:
                self.k.remove(i)
                print("Da Xoa")
                return
        print("Khong co san pham nay")
    def ktra(self):
        return len(self.k)==0
    def val(self):
        return f"Tong tien san pham la: {sum(i["price"] for i in self.k)}"
    def hienthi(self):
        if self.ktra():
            print("Khong co do")
            return
        for i, item in enumerate(self.k,1):
            print(f"{i}, {item["name"]}:{item["price"]}")
    def xoahet(self):
        self.k=[]
        print("Da Xoa Het San Pham") 
gio=ShoppingCart()
while True:
    print("2.Them | 3.Xoa | 4. Kiem tra trong | 5. Tong tien | 6. Hien thi | 7. Xoa het san pham")
    a=int(input())
    if a==2:
        name,price=input().split()
        gio.add(name,float(price))
    elif a==3:
        name=input()
        gio.de(name)
    elif a==4:
        if gio.ktra():
            print("Trong")
        else:
            print("CO DO")
    elif a==5:
        print(gio.val())
    elif a==6:
        gio.hienthi()
    elif a==7:
        gio.xoahet()
    e=input("BAN CO MUON THOAT KO?(yes/no)").strip().lower()
    if e=="yes":
        print("Bye")
        break
```

```python
class Rectangle:
    def __init__(self,w,h):
        self.w=w
        self.h=h
    def area(self):
        return self.w*self.h
    def perimeter(self):
        return 2*(self.w+self.h)
    def scale(self,x):
        self.w*=x
        self.h*=x
w,h,k=map(float,input().split())
kq=Rectangle(w,h)
kq.scale(k)
print(f"{kq.area():.2f}",end=" ")
print(f"{kq.perimeter():.2f}")
```

## W8A7

```python
class Fraction:
    def __init__(self,a,b):
        if b==0:
            raise ValueError("Mau so phai khac 0")
        if b<0:
            a=-a
            b=-b
        x,y=abs(a),abs(b)
        while y>0:
            x,y=y,x%y
        self.a=a//x
        self.b=b//x
    def add(self,new):
        c=self.a*new.b+self.b*new.a
        d=self.b*new.b
        return Fraction(c,d)
    def sub(self,new):
        c=self.a*new.b-self.b*new.a
        d=self.b*new.b
        return Fraction(c,d)
    def mul(self,new):
        c=self.a*new.a
        d=self.b*new.b
        return Fraction(c,d)
    def div(self,new):
        if new.a==0:
            raise ValueError("Khong the chia")
        c=self.a*new.b
        d=self.b*new.a
        return Fraction(c,d)
    def __str__(self):
        if self.b==1:
            return f"{self.a}"
        return f"{self.a}/{self.b}"
a,b,op,c,d=input().split()
f1=Fraction(int(a),int(b))
f2=Fraction(int(c),int(d))
if op=="+":
    print(f1.add(f2))
elif op=="-":
    print(f1.sub(f2))
elif op=="*":
    print(f1.mul(f2))
elif op=="/":
    print(f1.div(f2))
```

```python
class BankAccount:
    def __init__(self,owner,balance):
        self.o=owner
        self.b=balance
    def deposit(self,x):
        self.b+=x
        print("Nap thanh cong")
    def withdraw(self,x):
        if x<=self.b:
            self.b-=x
            print("Rut thanh cong")
            return
        print("Rut khong thanh cong")
    def sodu(self):
        return self.b
o,b=input().split()
kq=BankAccount(o,float(b))
n=int(input())
for i in range(n):
    a=input().split()
    if a[0]=="DEPOSIT":
        kq.deposit(float(a[1]))
    elif a[0]=="WITHDRAW":
        kq.withdraw(float(a[1]))
print(f"{kq.sodu():.2f}")
```

```python
class Student:
    def __init__(self,name):
        self.n=name
        self.tong=0
        self.dem=0
    def add_score(self,s,score):
        self.tong+=score
        self.dem+=1
    def avg(self):
        return self.tong/self.dem
    def rank(self):
        if self.avg()>=8:
            return "Excellent"
        if self.avg()>=6.5:
            return "Good"
        if self.avg()>=5:
            return "Average"
        else:
            return "Poor"
n=input()
a=int(input())
kq=Student(n)
for i in range(a):
    sub,sc=input().split()
    kq.add_score(sub,float(sc))
print(f"{n} {kq.avg():.2f} {kq.rank()}")
```

## W8A11

```python
import math
class Point3:
    def __init__(self,x,y,z):
        self.x=x
        self.y=y
        self.z=z
    def tru_vt(self,other):
        return [other.x-self.x,other.y-self.y,other.z-self.z]
def nhan_co_huong(u,v):
    return [u[1]*v[2]-u[2]*v[1],u[2]*v[0]-v[2]*u[0],u[0]*v[1]-u[1]*v[0]]
def nhan_vo_huong(u,v):
    return u[0]*v[0] + u[1]*v[1] + u[2]*v[2]
def do_dai(u):
    return (u[0]**2+u[1]**2+u[2]**2)**0.5
def goc(A,B,C,D):
    ab=A.tru_vt(B)
    bc=B.tru_vt(C)
    cd=C.tru_vt(D)
    n1=nhan_co_huong(ab,bc)
    n2=nhan_co_huong(bc,cd)
    tren=nhan_vo_huong(n1,n2)
    duoi=do_dai(n1)*do_dai(n2)
    if duoi==0:
        return 0
    cos_phi=abs(tren)/duoi
    rad=math.acos(cos_phi)
    do=math.degrees(rad)
    return do
k=[]
for _ in range(4):
    a=list(map(float,input().split()))
    kq=Point3(a[0],a[1],a[2])
    k.append(kq)
A,B,C,D=k[0],k[1],k[2],k[3]
phi=goc(A,B,C,D)
print(f"{phi:.2f}")
```

```python
class Triangle:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def dien_tich(self):
        if self.a+self.b<=self.c or self.a+self.b<=self.c or self.b+self.c<=self.a:
            return -1
        p=(self.a+self.b+self.c)/2
        m=(p*(p-self.a)*(p-self.b)*(p-self.c))**0.5
        return m
a,b,c=map(float,input().split())
kq=Triangle(a,b,c)
new=kq.dien_tich()
if new==-1:
    print("Invalid")
else:
    print(f"{new:.2f}")
```

```python
class Employee:
    def __init__(self,name,salary):
        self.n=name
        self.s=salary
    def income(self):
        return self.s
class Manager(Employee):
    def __init__(self,name,salary,bonus):
        super().__init__(name,salary)
        self.b=bonus
    def income(self):
        return self.s+self.b
a=input()
if a=="Employee":
    name,salary=input().split()
    kq=Employee(name,int(salary))
    print(f"{name} {kq.income()}")
elif a=="Manager":
    name,salary,bonus=input().split()
    kq=Manager(name,int(salary),int(bonus))
    print(f"{name} {kq.income()}")
```
