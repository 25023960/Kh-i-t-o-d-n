## W4A13

```python
def cu(n):
    c=0
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            c+=i
            if (n//i)!=i:
                c+=n//i
    return c+1
a,b=map(int,input().split())
if cu(a)==b and cu(b)==a:
    print("True")
else:
    print("False")
```
