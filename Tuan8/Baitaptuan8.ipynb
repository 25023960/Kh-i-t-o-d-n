## W7A1

```python
def bi(a,target,l,r):
    if l>r:
        return -1
    mid=(l+r)//2
    if a[mid]==target:
        return mid
    elif a[mid]<target:
        return bi(a,target,mid+1,r)
    else:
        return bi(a,target,l,mid-1)
a=list(map(int,input().split()))
target=int(input())
l=0
r=len(a)-1
print(bi(a,target,l,r))
```

## W7A3

### BUBBLE SORT

```python
def bb(a):
    for i in range(len(a)):
        k=False
        for j in range(len(a)-i-1):
            if a[j]>a[j+1]:
                a[j],a[j+1]=a[j+1],a[j]
                k=True
        if k==False:
            return a
a=list(map(int,input().split()))
print(bb(a))
```

### SELECTION SORT

```python
def se(a):
    l=len(a)
    for i in range(len(a)):
        mi=i
        for j in range(i+1,len(a)):
            if a[mi]>a[j]:
                mi=j
        a[i],a[mi]=a[mi],a[i]
    return a
a=list(map(int,input().split()))
print(se(a))
```

## W7A4

```python
def xh(a):
    c={}
    for i in a:
        c[i]=c.get(i,0)+1
    ma=0
    sl=0
    for key,value in c.items():
        if c[key]>sl:
            ma=key
            sl=value
    return ma,sl
a=list(map(int,input().split()))
print(xh(a))
```

## W7A5

```python
def so(a,c):
    k=0
    for i in range(len(a)-1):
        for j in range(i+1,len(a)):
            if (a[i]+a[j])%c==0:
                k+=1
    return k
a=list(map(int,input().split()))
c=int(input())
print(so(a,c))
```

## W7A7

```python
def d(a):
    dp=[1]*(len(a))
    for i in range(len(a)):
        for j in range(i):
            if a[i]>a[j]:
                dp[i]=max(dp[i],dp[j]+1)
    return max(dp)
u=input()
a=eval(u)
print(d(a))
```

## W7A8

```python
def q(a,c):
    k=[]
    for i in c:
        mi=float("inf")
        for l,r in a:
            if l<=i<=r:
                le=r-l+1
                if mi>le:
                    mi=le
        k.append(mi if mi!=float("inf") else -1)
    return k
m=input()
n=input()
a=eval(m)
c=eval(n)
print(q(a,c))
```

```python
def tim(a):
    k=[]
    for i in range(len(a)-1):
        dem=0
        for j in range(i+1,len(a)):
            if a[i]>a[j]:
                dem+=1
        k.append(dem)
    k.append(0)
    return k
u=input()
a=eval(u)
print(tim(a))
```

## W7A10


```python
def du(a,target):
    res=0
    for i in range(len(a)):
        for j in range(i,len(a)):
            if sum(a[i:j])%target>res:
                res=sum(a[i:j])%target
    return res
u=input()
target=int(input())
a=eval(u)
print(du(a,target))
```

## W7A13

```python
def hh(a):
    c={}
    for i in a:
        c[i]=c.get(i,0)+1
    l=0
    for i in c:
        if i+1 in c:
            if c[i+1]+c[i]>l:
                l=c[i+1]+c[i]
    return l
a=list(map(int,input().split()))
print(hh(a))
```
