## W7A19

```python
def me(a,b):
    i,j=0,0
    k=[]
    while i<len(a) and j<len(b):
        if a[i]<b[j]:
            k.append(a[i])
            i+=1
        else:
            k.append(b[j])
            j+=1
    return k+a[i:]+b[j:]
a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(me(a,b))
```

## W7A28

```python
def snt(n):
    dp=[True]*(n+1)
    dp[0]=dp[1]=False
    for i in range(2,int(n**0.5)+1):
        if dp[i]:
            for j in range(i*i,n+1,i):
                dp[j]=False
    return dp

n=int(input())
dem=0
ma=0
fid=0
p=snt(n)
for i in range(2,n+1):
    if not p[i]:
        dem+=1
    else:
        if ma<dem:
            ma=dem
            fid=i-dem
        dem=0
if ma<dem:
    ma=dem
    fid=n+1-dem
print(*(i for i in range(fid,fid+ma)))
```

## W7A30

```python
a=list(map(int,input().split()))
dp=[1]*(len(a))
trace=[-1]*(len(a))
for i in range(len(a)):
    for j in range(i):
        if a[i]>a[j] and dp[i]<dp[j]+1:
            dp[i]=dp[j]+1
            trace[i]=j
ma=max(dp)
last=dp.index(ma)
k=[]
while last!=-1:
    k.append(a[last])
    last=trace[last]
print(k[::-1])
```
