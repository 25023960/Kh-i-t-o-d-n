## W7A19

```python
def me(a,b):
    i,j=0,0
    k=[]
    while i<len(a) and j<len(b):
        if a[i]<b[j]:
            k.append(a[i])
            i+=1
        else:
            k.append(b[j])
            j+=1
    k.extend(a[i:])
    k.extend(b[j:])
    return k
a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(me(a,b))
```
