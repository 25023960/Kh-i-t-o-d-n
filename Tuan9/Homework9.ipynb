```python
class Cylinder:
    def __init__(self,r,h):
        self.bk=r
        self.cc=h
    def sbm(self):
        return 2*3.14*(self.bk)**2+2*3.14*(self.bk)*(self.cc)
    def tt(self):
        return (self.bk)**2*3.14*(self.cc)
    def __str__(self):
        return f"{self.bk},{self.cc}"
r,h=map(float,input().split())
cy1=Cylinder(r,h)
print(f"{cy1.sbm():.2f}")
print(f"{cy1.tt():.2f}")
print(cy1.cc)
```

```python
class MyDate:
    def __init__(self,dd,mm,yy):
        self.d=dd
        self.m=mm
        self.y=yy
    def cV(self):
        maxday=-1
        match self.m:
            case 1|3|5|7|8|10|12:
                maxday=31
            case 4|6|8|10|12:
                maxday=30
            case 2:
                if (self.y%4==0 and self.y%100!=0) or (self.y%400==0):
                    maxday=29
                else:
                    maxday=28
        return 1<=self.d<=maxday and self.y>0
    def nextDay(self):
        if (self.cV()):
            day1=MyDate(self.d+1,self.m,self.y)
            if day1.cV():
                return day1
            day1=MyDate(1,self.m+1,self.y)
            if day1.cV():
                return day1
            day1=MyDate(1,1,self.y+1)
            if day1.cV():
                return day1
    def __str__(self):
        return f"{self.d:02d}/{self.m:02d}/{self.y}"
d,m,y=map(int,input().split("/"))
day1=MyDate(d,m,y)
day2=day1.nextDay()
print(day2)
```

```python
class Shape:
    def area(self):
        pass
class Rectangle(Shape):
    def __init__(self,w,h):
        self.dai=h
        self.rong=w
    def area(self):
        return self.dai*self.rong
class Circle(Shape):
    def __init__(self,r):
        self.r=r
    def area(self):
        return 3.14*self.r**2
a=input()
if a=="Rectangle":
    w,h=map(float,input().split())
    cn=Rectangle(w,h)
    print(f"{cn.area():.2f}")
if a=="Circle":
    r=float(input())
    tr=Circle(r)
    print(f"{tr.area():.2f}")
```

```python
class Printer:
    def print_value(self,x):
        print(x)
class BinaryPrinter(Printer):
    def print_value(self,x):
        k=""
        while x>0:
            if x%2==0:
                k="0"+k
            else:
                k="1"+k
            x//=2
        print(int(k))
a=int(input()) 
m=BinaryPrinter()
m.print_value(a)
```

```python
class Stack:
    def __init__(self):
        self.k=[]
    def push(self,x):
        self.k.append(x)
    def pop(self):
        if len(self.k)>0:
            self.k.pop()
    def top(self):
        return self.k[-1]
n=int(input())
kq=Stack()
for i in range(n):
    a=input().split()
    if a[0]=="PUSH":
        kq.push(int(a[1]))
    elif a[0]=="POP":
        kq.pop()
    elif a[0]=="TOP":
        print(kq.top())
```
